## In this project, I will attempt to cluster university into Private and Public

This data has a label column but will be dropped for the Unsupervised work. 

```python
# Import the libraries/packages

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Import the data

datala = pd.read_csv("College_Data", index_col=0)
```

```python
data = datala.drop("Private", axis=1)
```


## The Metadata

* Accept Number of applications accepted

```python
# Since the data is clean enough for the model. 
# From the requirement, there are two clusters private or public
# Instantiate the model

kmeans = KMeans(n_clusters=2)
```
